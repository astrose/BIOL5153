```python
my_dna="ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT"
```

```python
#amount of A's present
a_count=my_dna.count('A')
```

```python
#amountof T's present 
t_count=my_dna.count('T')
```

```python
#Printing A count
print("The A count is" + " " +str(a_count)+".")
```

```python
#Printing T count
print("The T count is" + " " +str(t_count)+".")
```

```python
#Printing the total count
print("The total count is"+" "+ str(a_count + t_count)+".")
```

```python
#Calculate the length of my_dna
length=len(my_dna)
print(length)
```

```python
#Answer: AT content
print("The AT content is"+" "+str((a_count+t_count)/length)+".")
```
